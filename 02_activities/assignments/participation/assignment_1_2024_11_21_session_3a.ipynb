```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

np.random.seed(613)
x = np.arange(50) # a range of random numbers from 0 to 49
y1 = np.random.randint(0,100,50) # a range of 50 random numbers from 0 to 99
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize=(5,3)) # size of graph is 5X3 inches 
ax.plot(x,y1, y2) # plot as scatter graph
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1)
ax.plot(x, y2)

# this is another way to do it and better if you're going to format each line separately
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = "Person 1") # labelling x, y1
ax.plot(x, y2, label = "Person 2") # labelling x, y2
ax.legend(loc="lower right") # adding a legend in the lower right corner
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = "Person 1") # labelling x, y1
ax.plot(x, y2, label = "Person 2") # labelling x, y2
ax.legend(loc="lower right", # adding a legend in the lower right corner
          frameon = True, # add frame around the legend
          fontsize = 12, # change font size
          ncol = 2, # specify number of columns
          shadow = True) # add shadow around the frame
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = "Person 1") # labelling x, y1
ax.plot(x, y2, label = "Person 2") # labelling x, y2
ax.legend(loc="upper left", # adding a legend in the lower right corner
          bbox_to_anchor = (1,1)) # where we want to pin our legend box 

# for this example, the upper left corner of our legend is pinned to (1,1) position
# (1,1) is like 100% of the horizontal and vertical lines
# if it's (1,0.5), the legend will move to the 50% position of the vertical line
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1") # labelling x, y1
ax.scatter(x, y2, label = "Person 2") # labelling x, y2
ax.legend(loc="lower right") # adding a legend in the lower right corner
ax.text(10, 95, "This value is important!") # adding text to our graph, positioned at x=10, y=95
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "Person 1") 
ax.scatter(x, y2, label = "Person 2") 
ax.legend(loc="lower right") 
ax.text(10, 95, "This value is important!",
        ha = "center", # alignment of the text
        color = "red", # colour of the text
        size = 20) # size of the text
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # empty graph with x-axis going from 0-10 and y-axis going from 0-10

ax.text(1, 5, ". Data: (1,5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)

# adding text and points to the graph in different positions (the coordinates in pink)
# transform is referring to the coordinates of where to place the text
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.annotate("This is important!", xy = (10, 95),
            xytext = (20,94),
            arrowprops = dict(facecolor = 'black', shrink = 0.05)) # adding an arrow in black, shrinking away from text

# xy = (10,95) refers to where we want the arrow to point
# xytext is referring to where the text starts
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.annotate("This is important!", xy = (10, 95),
            xytext = (20,94),
            arrowprops = dict(facecolor = 'hotpink', arrowstyle = "wedge"))

# changing arrow colour and style
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.yaxis.set_major_locator(plt.NullLocator()) # locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) # formatter to remove only labels
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # we only want 3 tick marks on the x-axis
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # adding tick marks on x-axis at an interval of 5

# can do this with y-axis too, just change to ax.yaxis
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, # rotating numbers on x-axis by 45 degrees
           ha='right', # alignment of text
           font = 'serif',
           fontsize = 12)
```

```python
font1 = {'family':'serif','color':'indigo'}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = 'lower right')

plt.xticks(rotation=45,
           ha='right',
           font = 'serif',
           fontsize = 12)

plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
# adding label to x-axis
```
