```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50) # 50 random numbers from 0 to 99

fig, ax = plt.subplots(figsize=(5,3)) # size of graph is 5X3 inches 
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # histograms only count the occurrences of each value of "y"
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
font1 = {'family': 'sans-serif', 'color':'blue', 'size':20}
font2= {'family': 'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict= font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
            marker = '*',
            color = 'indigo')

# using star shape as markers in indigo colour
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = '*',
            color = 'indigo',
            linestyle = '--',
            linewidth = 2,
            markeredgecolor = '#FA9359', # orange coloured edges on the stars
            markerfacecolor = '#000000') # black colour in the middle of the stars

#ax.grid() # adding horizontal and vertical grid lines
ax.grid(axis = 'y') # adding horizontal grid lines
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker = 'P',
            color = 'darkturquoise',
            linestyle = ':',
            linewidth = 2,
            markeredgecolor = 'gold', # orange coloured edges on the stars
            markerfacecolor = 'thistle') # black colour in the middle of the stars

#ax.grid() # adding horizontal and vertical grid lines
ax.grid(axis = 'x') # adding horizontal grid lines
```
