```python
# importing "tips" dataset from seaborn
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data=tips,       # choose our dataset
           x='total_bill',  # define our x variable
           y='tip')         # define our y variable
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
                      x='total_bill',
                      y='tip',
                      color = 'hotpink',
                      linestyle = '--',
                      linewidth = 3,
                      marker = 'o',
                      markerfacecolor = 'indigo')

tipgraph.set(title='Tips vs. Total Bill',
            xlabel='Total Bill ($)',
            ylabel='Tip Amount ($)')
```

```python
tipgraph = sns.scatterplot(data = tips,
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time', # change style of the points based on the time variable
                           hue = 'day', # change hue of the points based on the day variable
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title = 'Tip vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# plotting everything in our dataset with different colours
# good for seeing potential trends when doing EDA
sns.pairplot(data = tips, hue = 'day')
```

```python
# tips vs total_bill for each day based on sex

daysplot = sns.relplot(data = tips,
                       x = "total_bill",
                       y = "tip",
                       hue = "sex", # colour the points by sex
                       col = "day", # columns should be based on our day variable
                       kind = "scatter", # kind of graph being made (scatter)
                       col_wrap = 2) # 2 graphs per column
```

```python
# using plotly to make visualizations
import plotly.graph_objects as go 

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array ([100, 180, 240, 99, 220])
```

```python
# defining our figure as "graph"
graph = go.Figure()

# choose plot type (bar graph) and adding it to our graph object
graph.add_trace(go.Bar(x=x1, y=y1))

# adding titles and labels
graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score")

# if you hover over the bars, you can see more information
# you can save as a png file
# you can zoom in
# lots of dynamic and interative features
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x = x1,
    y = y1,
    mode = 'markers',
    marker = dict( # changing the markers on the graph
        size = 15,
        color = 'hotpink',
        opacity = 1,
        line = dict(width = 5, # changing the line around the markers
                    color = 'purple')
    )
))

graph.update_layout(title = "interactive Plot",
                    xaxis_title = "Pirates",
                    yaxis_title = "Scores",
                    width = 500, # changing the width and height of the graph
                    height = 500)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", 
                 on_bad_lines = 'skip')

df
```

```python
text = " ".join(each for each in df.quote)
# for every quote in that column, we want to join them together with a space
# this creates a long string

wordcloud = WordCloud(background_color = 'white',
                      colormap = 'inferno').generate(text)
# inferno is a colour scheme in wordcloud

fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,
          interpolation = 'bilinear') # bilinear makes the data smoother and fit better
ax.axis("off") # turning off the axes and grid lines
```

```python
# making venn diagrams

from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])

diagram = venn2([A, B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue', 'red'),
                alpha = 0.5) # alpha is transparency

diagram.get_label_by_id("10").set_text("\n".join(A - B))
# 10 refers to specific positioning
# output all the labels from set A, separating each by a new line
diagram.get_label_by_id("11").set_text("\n".join(A & B))
# output all labels from set A and B
diagram.get_label_by_id("01").set_text("\n".join(B - A))
# output only labels from set B, separating each by a new line
```
