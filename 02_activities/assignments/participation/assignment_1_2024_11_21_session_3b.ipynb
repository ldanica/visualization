```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

# creating our data
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2, 
                        nrows=1, 
                        figsize=(7, 3))

ax1.scatter(x1, y1)
ax2.bar(x2, y2)
fig.show()
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2, #number of columns is 2
                        nrows=1, # number of rows is 1
                        figsize=(9, 3))

ax1.scatter(x1, y1, marker = 'o', color = 'red')
ax2.bar(x2, y2, color = 'darkgreen')
fig.show()

ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
ax2.yaxis.set_major_locator(plt.MultipleLocator(20))    

# plotting 2 graphs next to each other
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
# someaxes labels ax1, ax2, and ax3
# we want ax1 to be top left, ax2 to be bottom left, and ax3 to go from top right to bottom right

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
# then we input and graph our data
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label',fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize = 18)

# when labelling gets complicated
# 'A Big Label' is hidden but if you change figsize, it'll show up
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label',fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize = 18)

# adding layout = constrained so that all the subplots fit neatly within the figsize
```
